```python
import pandas as pd

# List of URLs to scrape
urls = [
    'https://en.wikipedia.org/wiki/Bosch_(company)',
    'https://en.wikipedia.org/wiki/Samsung_SDI',
    'https://en.wikipedia.org/wiki/BYD_Company',
    'https://en.wikipedia.org/wiki/CATL',
    'https://en.wikipedia.org/wiki/LG_Chem',
    'https://en.wikipedia.org/wiki/LG_Energy_Solution',
    'https://en.wikipedia.org/wiki/Toshiba',
    'https://en.wikipedia.org/wiki/Sony',
    'https://en.wikipedia.org/wiki/Sanyo',
    'https://en.wikipedia.org/wiki/Nissan',
    'https://en.wikipedia.org/wiki/Hitachi',
    'https://en.wikipedia.org/wiki/Hyundai_Motor_Company',
    'https://en.wikipedia.org/wiki/GS_Yuasa',
    'https://en.wikipedia.org/wiki/Panasonic'
]

# List to hold all successfully scraped data
scraped_data_collection = []

# Initialize the scraper with specific instructions for the data format
for url in urls:
    scraper = SmartScraperGraph(
        prompt="Scrape the website and extract the company_name, Company type, Founded, Founder, Headquarters, Area_served, Key people, Products, Number of employees, Revenue, Operating income, Net income, Total assets, year, Total_equity, Website",
        source=url,
        config=graph_config
    )
    # Execute the scraping process
    data = scraper.run()
    scraped_data_collection.append(data)

# Create a DataFrame from the scraped data
df = pd.DataFrame(scraped_data_collection)

# Display the DataFrame
print(df)
```
